```python
ratings = df_final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

observation

 1. When Rating is between 4.5 to 4.9 = Excellent
 2. When Rating is between 4.0 to 4.4 = Very Good
 3. When Rating is between 3.5 to 3.9 = Good
 4. When Rating is between 2.5 to 3.4 = Average
 5. When Rating is Between 1.8 to 2.4 = Poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sb.barplot(x="Aggregate rating",y="Rating Count", data=ratings)
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sb.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',
           data=ratings,palette=['pink','red','orange','yellow','green','blue'])
```

Observation

 1. not rated count is very high
 2. Maximum number of rating are between 2.5 to 4.4

```python
# Count plot 

sb.countplot(x = "Rating color", data = ratings, palette=['pink','red','orange','yellow','green','blue'])
```

```python
# Find the country name thats give zero rating.

df_final[df_final["Aggregate rating"] ==0].groupby('Country').size().reset_index()
```
